# Import Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

# Loading Dataset

```python
gold_data = pd.read_csv('gld_price_data.csv')
gold_data
```

```python
gold_data.info()
```

```python
gold_data.describe()
```

```python
gold_data['Date'] = pd.to_datetime(gold_data['Date'], errors='coerce')
```

```python
corr = gold_data.drop(columns=['Date']).corr()
```

```python
corr = gold_data.corr()
corr
```

```python
sns.heatmap(corr,cbar=True, square=True,annot = True, cmap = 'Blues')
plt.show()
```

```python
# Correlation values of gold
print(corr['GLD'])
```

```python
# checking the distribution of gold price
sns.distplot(gold_data['GLD'], color = 'black')
plt.show()
```

# Split the feature and target

```python
x = gold_data.drop(['Date', 'GLD'], axis = 1)
y = gold_data['GLD']
```

```python
x,y
```

# Train and Test the Model

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=3)
```

```python
rg = RandomForestRegressor(n_estimators=100)
```

```python
rg.fit(x_train, y_train)
```

```python
test_data_pred = rg.predict(x_test)
test_data_pred
```

```python
error_score = metrics.r2_score(y_test, test_data_pred)
error_score
```

# Comapre the actual values and predicted value in a plot

```python
y_test = list(y_test)
```

```python
plt.plot(y_test, color = 'red', label = 'Actual Value')
plt.plot(test_data_pred, color = 'black', label = 'Predicted Value')
plt.title('Actual Price vs Predicted Value')
plt.xlabel('Number of Values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
